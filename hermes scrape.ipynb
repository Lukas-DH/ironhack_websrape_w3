```python
import time
df_final = pd.DataFrame()


for cats in catagories:
    i = 0
    df_shape = 1 , 1
    while df_shape != df_final.shape:  
        df_shape = df_final.shape
        url = f'https://bck.hermes.com/product?urlParams=fh_view_size=36%26country=uk%26fh_location=--%252Fcategories%253C%257B{cats.lower()}%257D%26fh_start_index={36*1 if i<2 else 36*(i-1)}&locale=uk_en&category={cats}&sort=relevance&offset={36*1 if i<2 else 36*(i-1)}'
        results = requests.get(url).json()
        df = pd.json_normalize(results['products'])
        df['pimCode'] = cats
        df_final = df_final.append(df)
        time.sleep(0.1)
        i += 1
    print(cats)
    print(df_final.shape)
```

```python
def flatten(data,list_col):
    for column in list_col:
        flattened=pd.DataFrame(dict(data[column])).T
        columns=[str(col) for col in flattened.columns]
        flattened.columns=[column + '_' + col for col in columns]
        data=pd.concat([data,flattened], axis=1)
        data=data.drop(column,axis=1)
    return data
```
